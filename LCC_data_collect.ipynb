# Skytrax

```python
wizz = "https://www.airlinequality.com/airline-reviews/wizz-air/page/@/?sortby=post_date%3ADesc&pagesize=100" # 16
ryan = "https://www.airlinequality.com/airline-reviews/ryanair/page/@/?sortby=post_date%3ADesc&pagesize=100" # 24
airasia = "https://www.airlinequality.com/airline-reviews/airasia/page/@/?sortby=post_date%3ADesc&pagesize=100" # 11
scoot = "https://www.airlinequality.com/airline-reviews/scoot/page/@/?sortby=post_date%3ADesc&pagesize=100" # 8
pages = [16, 24, 11, 8]
urls = [wizz, ryan, airasia, scoot]
```

```python
from seleniumbase import Driver
from bs4 import BeautifulSoup
```

```python
driver = Driver(browser="Chrome", uc=True)
```

```python
from time import sleep
from random import randint
reviews = []
for url, page in list(zip(urls, pages)):
    review = []
    for p in range(1, page):
        new_url = url.replace("@", str(p))
        driver.get(url=new_url)
        sleep(randint(5,7))
        doc = BeautifulSoup(driver.get_page_source(), "html.parser")
        review.extend([w.text for w in doc.find_all(class_="text_content")])
    reviews.append(review)
```

```python
label = ["Wizz"] * len(reviews[0])
label += ["Ryan"] * len(reviews[1])
label += ["AirAsia"] * len(reviews[2])
label += ["Scoot"] * len(reviews[3])
label
```

```python
import pandas as pd
review_all = []
for r in reviews:
    review_all.extend(r)
df = pd.DataFrame({"Airline": label, "Review": review_all})
```

```python
df.head()
```

```python
df.to_csv("Skytrax.csv")
```

# Trustpilot

```python
urls2 = ["https://www.trustpilot.com/review/www.wizzair.com",
"https://www.trustpilot.com/review/www.ryanair.com",
"https://www.trustpilot.com/review/www.airasia.com",
"https://www.trustpilot.com/review/flyscoot.com"] # 27
```

```python
next_button = driver.find_element(By.NAME, "pagination-button-next")
next_button.click()
```

```python
doc = BeautifulSoup(driver.get_page_source(), "html.parser")
[w.p.text for w in doc.find_all(class_="styles_reviewContent__tuXiN")]
```

```python
driver = Driver(browser="Chrome", uc=True)
```

```python
from selenium.common.exceptions import ElementClickInterceptedException, NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
reviews2 = []
for url in urls2:
    review2 = []
    driver.get(url=url)
    sleep(randint(3,5))
    i = 0
    while i < 100:
        doc = BeautifulSoup(driver.get_page_source(), "html.parser")
        review2.extend([w.p.text for w in doc.find_all(class_="styles_reviewContent__tuXiN") if w.p])
        clicked = False
        for attempt in range(3):
            try:
                next_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.NAME, "pagination-button-next"))
                )
                if not next_button.is_enabled():
                    break
                driver.execute_script("arguments[0].click();", next_button)
                clicked = True
                break
            except StaleElementReferenceException:
                sleep(1)
                continue
            except (ElementClickInterceptedException, NoSuchElementException):
                break
        i += 1
        if not clicked:
            break
        
        sleep(randint(3, 4))
    reviews2.append(review2)
```

```python
reviews2_no_dup = []
for r in reviews2:
    reviews2_no_dup.append(list(set(r)))
```

```python
len(reviews2_no_dup[3])
```

```python
label2 = ["Wizz"] * len(reviews2_no_dup[0])
label2 += ["Ryan"] * len(reviews2_no_dup[1])
label2 += ["AirAsia"] * len(reviews2_no_dup[2])
label2 += ["Scoot"] * len(reviews2_no_dup[3])
label2
```

```python
review2_all = []
for r in reviews2_no_dup:
    review2_all.extend(r)
df2 = pd.DataFrame({"Airline": label2, "Review": review2_all})
```

```python
df2.to_csv("Trustpilot.csv")
```
